```python
#Question 5
#What are the 5 most frequently mentioned ingredients?

#opening URL
import urllib.request
book_url = 'https://www.gutenberg.org/cache/epub/69346/pg69346.txt'

book_text = urllib.request.urlopen(book_url) #open URL as "book_text"
book_text = book_text.read()                 #returns all bytes from "book_text"
book_text = book_text.decode("utf-8")        #decode as UTF-8

#pre-processing English
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
doc = nlp(book_text) #tokenization & lemmatization using spaCy
from collections import Counter

# noun tokens that arent stop words or punctuations
nouns = [token.text
         for token in doc
         if (not token.is_stop and
             not token.is_punct and
             token.pos_ == "NOUN")]

# fifteen most common nouns
noun_freq = Counter(nouns)
common_nouns = noun_freq.most_common(15)
print(common_nouns)
```

```python
#Question 4
#open the URL and read the content
import urllib.request
book_url = 'https://www.gutenberg.org/cache/epub/69346/pg69346.txt'

book_text = urllib.request.urlopen(book_url) #open URL as "book_text"
book_text = book_text.read()                 #returns all bytes from "book_text"
book_text = book_text.decode("utf-8")        #decode as UTF-8

#pre-processing English
import spacy
import en_core_web_sm
nlp = en_core_web_sm.load()
doc = nlp(book_text) #tokenization & lemmatization using spaCy

#Are the recipes vegan or vegetarian?

#regex used to check if certain words are in the book text
import re
cheese = re.findall('cheese', book_text, flags=re.IGNORECASE)
print(f'Cheese mentioned: {len(cheese)}')
egg = re.findall('egg*', book_text, flags=re.IGNORECASE)
print(f'Egg mentioned: {len(egg)}')

#regular expressions used to check if certain words from not_vegan_list are present in the book text
not_vegan_list = ["cheese", "egg", "milk", "butter"]

#re.compile(pattern, repl, string):. We can combine a regular expression pattern into pattern objects, which can be used for pattern matching
if re.compile('|'.join(not_vegan_list),re.IGNORECASE).search(book_text): #re.IGNORECASE makes the search case-insensitive
    print("The recipes are vegetarian") # Do Something if word is present
else:
    print("The recipes are vegan") # Do Something else if word is not present

#regex to check if the word fish in the text
fish = re.findall('\\bfish\\b', book_text, flags=re.IGNORECASE)
print(f'Fish mentioned: {len(fish)}')

#possible solution, install numpy and pandas
import numpy as np
import pandas as pd

#split the book by recepies
splitted_list = book_text.split("INDEX")
print(splitted_list)

#defined a regular expression to split the string
pattern = r"\r\n\r\n+\r\n" #there are 2 empty lines after each recipe, then a line of text, then empty line

#used re.split() to split the string based on the regex pattern
result = re.split(pattern, book_text)
print(len(result))

#go through recepies (chosen manually)
start_index = 28
end_index = len(result) - 31

#create 2 lists
vegetarian = ["egg", "milk", "butter", "Beurre", "Béchamel", "cheese"]
fishy = ["lobster", "fish", "shrimp"]

#create empty lists
titles = []
list_is_vegan = []
list_is_vegetarian = []
#create loop to iterate over all recepies
for i in range(start_index, end_index + 1): #+1 to include the end_index, otherwise it won't include it
    is_vegan_i = False
    is_vegetarian_i = False
#text that belong to each recipe
    recep_text = result[i]
    #extract title from the text to understand what recipe
    titles.append(recep_text.split("\r\n\r\n")[0])
    # check if there is fish
    is_fishy = re.compile("|".join(fishy), re.IGNORECASE).search(recep_text)
    if is_fishy is None:
        # if there is no fish, it is vegeterian
        is_vegetarian_i = True
        # check if there is egg, butter, milk, etc.
        is_vegetarian = re.compile("|".join(vegetarian), re.IGNORECASE).search(recep_text)
        if is_vegetarian is None:
            # if don't find these words, we assume that is also vegan
            is_vegan_i = True
            #we go through loops and we save the result somewhere
    list_is_vegan.append(is_vegan_i)
    list_is_vegetarian.append(is_vegetarian_i)

            #create a DataFrame and filter data by title

df_result = pd.DataFrame(
    {"title": titles, "is_vegan": list_is_vegan, "is_vegetarian": list_is_vegetarian}
)
df_result
```
